# Checking Inverted Index

The following code is used to load and display an inverted index.

```python
from Data.saveIIndex import invDS
from Data.saveIIndex.parsing import parse
from Data.doc_data import doc_file_pb2 as dpb
from wordlexicon import lexicon_list
```

```python
inverted_index = invDS.load_from_binary_file("Data/inverted_index/invindex_82.bin")
```

```python
inverted_index.wordelems[:20]
```

```python
for i, wordelem in enumerate(inverted_index.wordelems):
    if len(wordelem) != 0:
        print(lexicon_list[i + 82*500], ": ", end = "")
        for eachdoc in wordelem[:5]:
            print(" (", end = "")
            print(eachdoc.doc_id, end = "-")
            print(eachdoc.hits, end="-")
            print(eachdoc.titlebool, end="-")
            print(eachdoc.importance, end="")
            print(") ", end = "")
        print()
```

```python
def get_eachdocdata(doc_id):
    doc_data = dpb.DocData()
    with open(f"Data/doc_data/docdata_{doc_id//10000}.pb", 'rb') as f:
        filecontent = f.read()
    doc_data.ParseFromString(filecontent)
    return doc_data.eachdoc[doc_id%10000]
```
